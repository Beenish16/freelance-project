```python
import tkinter as tk
from tkinter import ttk, messagebox
import sqlite3
from datetime import datetime, timedelta

# Database connection
conn = sqlite3.connect('library.db')
cursor = conn.cursor()

# Create tables if they don't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS books (
        id INTEGER PRIMARY KEY,
        title TEXT,
        author TEXT,
        quantity INTEGER
    )
''')
cursor.execute('''
    CREATE TABLE IF NOT EXISTS issued_books (
        id INTEGER PRIMARY KEY,
        book_id INTEGER,
        student_name TEXT,
        issue_date TEXT,
        return_date TEXT,
        FOREIGN KEY (book_id) REFERENCES books (id)
    )
''')
cursor.execute('''
    CREATE TABLE IF NOT EXISTS admins (
        id INTEGER PRIMARY KEY,
        username TEXT,
        password TEXT
    )
''')
conn.commit()

class LibraryManagementSystem:
    def _init_(self, root):
        self.root = root
        self.root.title("Library Management System")
        self.current_admin = None  # To store the current logged-in admin

        # Create and configure widgets...
        self.create_widgets()

    def create_widgets(self):
        # Create and configure your widgets (labels, buttons, entry fields, etc.)
        # Place them in the window using grid or pack geometry managers.

        # Sign In
        ttk.Label(self.root, text="Username:").grid(row=0, column=0, padx=10, pady=10)
        self.username_entry = ttk.Entry(self.root)
        self.username_entry.grid(row=0, column=1, padx=10, pady=10)

        ttk.Label(self.root, text="Password:").grid(row=1, column=0, padx=10, pady=10)
        self.password_entry = ttk.Entry(self.root, show="*")
        self.password_entry.grid(row=1, column=1, padx=10, pady=10)

        sign_in_button = ttk.Button(self.root, text="Sign In", command=self.sign_in)
        sign_in_button.grid(row=2, column=0, columnspan=2, pady=10)

        # Sign Out
        sign_out_button = ttk.Button(self.root, text="Sign Out", command=self.sign_out, state=tk.DISABLED)
        sign_out_button.grid(row=3, column=0, columnspan=2, pady=10)
        self.sign_out_button = sign_out_button

        # Example: Add Books
        ttk.Label(self.root, text="Title:").grid(row=4, column=0, padx=10, pady=10)
        self.title_entry = ttk.Entry(self.root)
        self.title_entry.grid(row=4, column=1, padx=10, pady=10)

        ttk.Label(self.root, text="Author:").grid(row=5, column=0, padx=10, pady=10)
        self.author_entry = ttk.Entry(self.root)
        self.author_entry.grid(row=5, column=1, padx=10, pady=10)

        ttk.Label(self.root, text="Quantity:").grid(row=6, column=0, padx=10, pady=10)
        self.quantity_entry = ttk.Entry(self.root)
        self.quantity_entry.grid(row=6, column=1, padx=10, pady=10)

        add_book_button = ttk.Button(self.root, text="Add Book", command=self.add_book, state=tk.DISABLED)
        add_book_button.grid(row=7, column=0, columnspan=2, pady=10)
        self.add_book_button = add_book_button

        # Repeat similar code for other functionalities...

    def sign_in(self):
        # Check the username and password against the database
        username = self.username_entry.get()
        password = self.password_entry.get()

        cursor.execute('SELECT * FROM admins WHERE username = ? AND password = ?', (username, password))
        admin = cursor.fetchone()

        if admin:
            self.current_admin = admin
            messagebox.showinfo("Success", "Sign in successful.")
            self.username_entry.delete(0, tk.END)
            self.password_entry.delete(0, tk.END)

            # Enable features after sign-in
            self.add_book_button["state"] = tk.NORMAL
            self.sign_out_button["state"] = tk.NORMAL
        else:
            messagebox.showerror("Error", "Invalid username or password.")

    def sign_out(self):
        self.current_admin = None
        messagebox.showinfo("Success", "Sign out successful.")

        # Disable features after sign-out
        self.add_book_button["state"] = tk.DISABLED
        self.sign_out_button["state"] = tk.DISABLED

    def add_book(self):
        # Get data from entry fields
        title = self.title_entry.get()
        author = self.author_entry.get()
        quantity = self.quantity_entry.get()

        # Insert into the database
        cursor.execute('INSERT INTO books (title, author, quantity) VALUES (?, ?, ?)', (title, author, quantity))
        conn.commit()

        messagebox.showinfo("Success", "Book added successfully.")

    # Implement other functionalities (issue_book, edit_book, return_book, delete_book, search_books, show_books)

    def run(self):
        self.root.mainloop()

if __name__ == "_main_":
    root = tk.Tk()
    app = LibraryManagementSystem(root)
    app.run()
```
